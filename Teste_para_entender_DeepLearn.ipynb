```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Dados de entrada (exemplo simples)
# X = características de entrada, Y = rótulos de saída
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Entradas
Y = np.array([[0], [1], [1], [0]])  # Saídas esperadas (XOR)

# Criando o modelo sequencial
model = Sequential()

# Adicionando uma camada oculta com 4 neurônios e função de ativação 'relu'
model.add(Dense(4, input_dim=2, activation='relu'))

# Adicionando a camada de saída com 1 neurônio e função de ativação 'sigmoid' (para classificação binária)
model.add(Dense(1, activation='sigmoid'))

# Compilando o modelo (usando a função de perda 'binary_crossentropy' para problemas de classificação binária)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Treinando o modelo (a retropropagação ocorre durante esse processo)
model.fit(X, Y, epochs=800, verbose=1)

# Avaliando o modelo
loss, accuracy = model.evaluate(X, Y)
print(f'Perda: {loss}, Acurácia: {accuracy}')

# Fazendo previsões
predictions = model.predict(X)
print(f'Previsões: {predictions}')
```
